```python
import random

#Create 5 element list. Name it thrownDiceNumbers.
#Pick randomly 5 numbers and assign them to the list.
thrownDiceNumbers = []
for i in range(5):
    thrownDiceNumbers.append(random.randint(1, 6))

#Print the whole list.
print(thrownDiceNumbers)

#Count the sum of the list values and print the result.
print("Sum of numbers:", sum(thrownDiceNumbers))

#Print the highest value of the list
print("Highest number:", max(thrownDiceNumbers))
```
