```python
num1 = 20
num2 = 20
type(num1)
type(num2)
print("num1 is",num1)
print("num2 is",num2)
print("")

#Test if number1 is equal to number2
if num1 == num2:
    print("num1 is equal to num2")

#Test if number1 is greater than number2.
if num1 > num2:
    print("num1 is greater than num2")

#Test if number1 is greater or equal to number2.
if num1 >= num2:
    print("num1 is greater or equal to num2")

#Test if number1 is NOT equal to number2.
if num1 != num2:
    print("num1 is not equal to num2")
```

```python
num1 = 10
num2 = 10
num3 = 10
type(num1)
type(num2)
type(num3)
print("num1 is",num1)
print("num2 is",num2)
print("num3 is",num3)
print("")

#Test if all numbers are equal.
if num1 == num2 and num2 == num3 and num1 == num3:
    print("all numbers are equal.")

#Test if number1 and number2 are equal OR number2 and number3 are equal.
if num1 == num2 or num2 == num3:
    print("num1 and num2 are equal OR num2 and num3 are equal.")

#Test if number1 is greater than number2 AND number1 is greater than number3.
if num1 > num2 and num1 > num3:
    print("num1 is greater than num2 and num3")

#Test if number1 is greater than number2. If not, test if number2 is greater than number3.
if num1 > num2:
    print("num1 is greater than num2")
elif num2 > num3:
    print("num2 is greater than num3")

#Test if number1 and number2 are equal. If not, test if number1 and number3 are equal.
if num1 == num2:
    print("num1 is equal to num2")
elif num1 == num3:
    print("num1 is equal to num3")
```
