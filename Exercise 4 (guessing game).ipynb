```python
guesses = 0
correctguess = "Valtteri"

while True:
    guess = input("Guess a name:")
    guesses += 1

    if guess.lower() == correctguess.lower():
        print("Congratulations, your guess was correct!")
        print("You guessed",guesses,"times.")
        break
    else:
        print("Your guess was incorrect, try again!")
```
